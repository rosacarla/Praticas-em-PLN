#### <p align='justify'>1) Utilize `Expressões Regulares` para remover de cada um dos textos do corpus o cabeçalho dos capítulos (e.g., "*Capítulo 1:*", "*Capítulo 2*:", "*Capítulo 3*:")</p>

```
# Importa módulo "re", necessário para lidar com expressões regulares.
import re
```

```
# Define função para remover cabeçalhos dos capítulos usando expressões regulares
def remover_cabecalhos(corpus):
    novo_corpus = []
    for texto in corpus:
        novo_texto = re.sub(r"Capítulo \d+: ", "", texto)
        novo_corpus.append(novo_texto)
    return novo_corpus

# Aplica a função no corpus
corpus_sem_cabecalhos = remover_cabecalhos(corpus)
```

```
# Imprime o resultado da remoção
for texto in corpus_sem_cabecalhos:
    print(texto)
```

#### <p align='justify'>2) Compute a quantidade de caracteres, quantidade de palavras, quantidade de palavras únicas e quantidade de sentenças de cada capítulo e apresente em um `DataFrame` da biblioteca pandas.</p>

```
# Importa biblioteca pandas para criar dataframe
import pandas as pd
```

```
# Define função para calcular as métricas de texto
def calcular_metricas(texto):
    # Quantidade de caracteres
    caracteres = len(texto)

    # Quantidade de palavras
    palavras = len(texto.split())

    # Quantidade de palavras únicas
    palavras_unicas = len(set(re.findall(r'\b\w+\b', texto.lower())))

    # Quantidade de sentenças
    sentencas = len(re.split(r'[.!?]', texto))

    return caracteres, palavras, palavras_unicas, sentencas
```

```
# Cria listas para armazenar métricas de cada capítulo
caracteres_por_capitulo = []
palavras_por_capitulo = []
palavras_unicas_por_capitulo = []
sentencas_por_capitulo = []
```

```
# Calcula métricas para cada capítulo
for texto in corpus:
    caracteres, palavras, palavras_unicas, sentencas = calcular_metricas(texto)
    caracteres_por_capitulo.append(caracteres)
    palavras_por_capitulo.append(palavras)
    palavras_unicas_por_capitulo.append(palavras_unicas)
    sentencas_por_capitulo.append(sentencas)
```

```
# Cria dataframe do pandas com as métricas
df_metricas = pd.DataFrame({
    'Capítulo': range(1, len(corpus) + 1),
    'Caracteres': caracteres_por_capitulo,
    'Palavras': palavras_por_capitulo,
    'Palavras Únicas': palavras_unicas_por_capitulo,
    'Sentenças': sentencas_por_capitulo
})

# Exibe o dataframe
print(df_metricas)
```

```
# Importa bibliotecas do NLTK
import nltk
nltk.download('punkt')
from nltk.metrics.distance import edit_distance
```

```
# Define função para encontrar e imprimir capítulos com erros de grafia
def contar_erros_grafia(corpus):
    # Contador de erros
    erros_por_capitulo = {}

    # Palavra correta
    palavra_correta = "hobbits"

    # Para cada capítulo no corpus
    for i, capitulo in enumerate(corpus, start=1):
        # Faz tokenização e remoção de pontuação
        palavras = nltk.word_tokenize(capitulo.lower())
        palavras = [p for p in palavras if p.isalnum()]

        # Verifica presença da palavra errada
        if "hobits" in palavras:
            # Conta o número de ocorrências da palavra errada no capítulo
            erros_por_capitulo[i] = palavras.count("hobits")

    return erros_por_capitulo

# Chamada da função para contar erros de grafia
erros_por_capitulo = contar_erros_grafia(corpus)
```

```
# Imprime número de erros por capítulo
for capitulo, erros in erros_por_capitulo.items():
    print(f"No Capítulo {capitulo} foram encontradas {erros} ocorrência(s) da palavra 'hobits'.")
```

```
# Instala modulo da biblioteca Levenshtein
!pip install Levenshtein
```

```
# Importa biblioteca que implementa algoritmo de distância de Levenshtein em Python
import Levenshtein
```

```
# Define função para encontrar erros de grafia em cada capítulo
def encontrar_erros_grafia(corpus):
    erros_de_grafia = []
    palavra_correta = "hobbits"
    limite_distancia = 2  # Limite de distância de Levenshtein para considerar um erro de grafia
    for i, texto in enumerate(corpus, start=1):
        if "hobits" in texto:
            palavras = texto.split()
            for j, palavra in enumerate(palavras):
                if Levenshtein.distance(palavra, palavra_correta) < limite_distancia:
                    erros_de_grafia.append((i, j + 1))  # Adiciona o capítulo e o índice da palavra
    return erros_de_grafia
```

```
# Encontra erros de grafia nos capítulos
erros = encontrar_erros_grafia(corpus)

# Imprime os resultados encontrados para "hobbit"
if len(erros) > 0:
    print("Erros de grafia encontrados nos seguintes capítulos:")
    for erro in erros:
        print(f"Capítulo {erro[0]}: Palavra na posição {erro[1]}")
else:
    print("Nenhum erro de grafia encontrado.")
```

```
# Importa biblioteca CountVectorizer do scikit-learn para criar vetores de palavra
from sklearn.feature_extraction.text import CountVectorizer
```

```
# Define as stop words
stop_words = ["uma", "um", "de", "para", "com", "no", "na", "e", "o", "a", "os", "as", "em", "por", "se", "do", "da", "dos", "das", "que"]

# Inicializa o CountVectorizer com as stop words
vectorizer = CountVectorizer(stop_words=stop_words)
```

```
# Aplica CountVectorizer no corpus
vetores_bow = vectorizer.fit_transform(corpus)

# Obtém palavras correspondentes aos índices no vocabulário
palavras_vocabulario = vectorizer.get_feature_names_out()
```

```
# Mostra vetores resultantes com suas palavras correspondentes
for i, vetor in enumerate(vetores_bow):
    print(f"Capítulo {i+1}:")
    for palavra_idx, frequencia in zip(vetor.indices, vetor.data):
        palavra = palavras_vocabulario[palavra_idx]
        print(f"    {palavra}: {frequencia}")
    print()
```
